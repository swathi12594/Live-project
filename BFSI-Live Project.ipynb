### Hypothesis Testing
### Gender

#### Null and Alternative Hypothesis
#### H0: Gender is impacting Loan status feature
#### H1: Gender is not impacting loan status feature

```python
fig = sns.countplot(x = "Loan_Status", data = df, hue = "Gender")
```

#### Implementing chi-square test

```python
G_tab = pd.crosstab(df.Loan_Status, df.Gender)
stats, p, dof, expected = chi2_contingency(G_tab)

print("P-value is",p)
```

#### Conclusion: From the graph we conclude that more loan approvals have been granted to Males. Hence Gender is impacting the decision of loan approval.
#### Failed to reject null hypothesis. Accepting Null Hypothesis 


#### From Chi square test P value is greater than 0.05, hence accept the null hypothesis, loan status is impacted by gender feature, both are dependent to each other

#### Loan amount, applicant income and coapplicant income these features are having skewed distribution, hence for detecting outliers I am using IQR method

```python
# Detecting outliers
def iqr_func(data):
    Q1 = np.percentile(data, 25, interpolation = 'midpoint') 
    Q3 = np.percentile(data, 75, interpolation = 'midpoint')
    IQR=Q3-Q1
    lo=Q1-1.5*IQR
    uo=Q3+1.5*IQR
    outlier=[]
    for x in data:
        if ((x> uo) or (x<lo)):
            outlier.append(x)
            
    print('outlier in the dataset are:', len(outlier ))
    
iqr_func(df['LoanAmount'])
iqr_func(df['ApplicantIncome'])
iqr_func(df['CoapplicantIncome'])
```
